```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('forestfires.csv')
df
```

```python
plt.scatter(df['month'], df['wind'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Month')
plt.ylabel('wind')

plt.show()
```

```python
#Bar chart with day against temprature
plt.bar(df['day'], df['temp'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Temprature')

# Adding the legends
plt.show()
```

```python
# histogram of total_bills
plt.hist(df['DMC'])
plt.title("Histogram")

plt.show()
```

```python
keys='FFMC','DMC','DC'
sizes=[df['FFMC'].sum(),df['DMC'].sum(),df['DC'].sum()]

# Plotting data on chart
plt.pie(sizes, labels=keys, autopct='%.0f%%')

# Add title to the chart
plt.title('Values of fire weather indices')

# Displaying chart
plt.show() 
```

```python
# draw lineplot
sns.lineplot(x="month", y="temp", data=df)
 
# setting the title 
plt.title('Line Chart')
 
plt.show()
```
